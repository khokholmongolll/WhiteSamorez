```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import math

x=np.array([ 2.05, 2.25, 2.45, 2.65, 2.95, 3.25, 3.85])
y=np.array([ 3360, 3930, 4470, 5050, 5880, 6720, 8320])

def least_squares(x, y):
    n = len(x)
    sumx = sum(x)/n
    sumy = sum(y)/n
    sumxy = sum(x*y)/n
    sumx_sq = sum(x**2)/n
    sumy_sq = sum(y**2)/n
    
    a = (sumxy - sumx * sumy) / (sumx_sq - sumx**2)
    b = sumy - a * sumx
    
    variance_a = (1/n**(1/2)) * (((sumy_sq - sumy**2)/(sumx_sq - sumx**2)) - b**2)**(1/2)
    variance_b = variance_a * (sumx_sq - sumx**2)**(1/2)
    
    return a, b, variance_a, variance_b

a, b, variance_a, variance_b = least_squares(x, y)
plt.scatter(x, y, color='blue')
plt.plot(x, a*x + b, color='red')
plt.xlabel('V, ')
plt.ylabel('kappa, /')
plt.grid(True)
plt.title('' )
plt.show()

print(a,b, variance_a)
```

```python
import numpy as np

#   ( /)
lambda_Na = 50.1
lambda_Cl = 76.3
lambda_H = 349.65
lambda_OH = 147.16  #  

#   ( /)
kappa1 = 10900e-6  # 0.0109 /
kappa2 = 1145e-6   # 0.001145 /
kappa3 = 1607e-6   # 0.001607 /

# 
C0 = 0.01  #  (  )
V0 = 70     #  ( )
C1 = 1.0    #  ( )

#     
V1 = -1.05  #  ( HCl)
V2 = 0.38   #  ( NaOH)
V3 = 1.06   #  ( NaOH)

#  [H+], [Cl-], [Na+], [OH-]   
def calculate_concentrations(Vt):
    if Vt < 0:
        #  HCl: [Cl-] = |Vt|*C1 / V0, [Na+] = 0
        Cl = abs(Vt) * C1 / V0
        Na = 0.0
        OH = 0.0  #    [OH-]  
    else:
        #  NaOH: [Na+] = Vt*C1 / V0, [Cl-] = 0
        Na = Vt * C1 / V0
        Cl = 0.0
        OH = Na  #    [OH-]  [Na+]
    return Na, Cl, OH

#   1 (V1 = -1.05 , pH1 = 1.72)
Na1, Cl1, OH1 = calculate_concentrations(V1)
H1 = 10**(-1.72)  # [H+] = 10^(-pH)

#   2 (V2 = 0.38 , pH2 = 6.05)
Na2, Cl2, OH2 = calculate_concentrations(V2)
H2 = 10**(-6.05)

#   3 (V3 = 1.06 , pH3 = 9.07)
Na3, Cl3, OH3 = calculate_concentrations(V3)
H3 = 10**(-9.07)

#    (b =  -  )
b1 = kappa1 - (lambda_H * H1 + lambda_Cl * Cl1)
b2 = kappa2 - (lambda_H * H2 + lambda_Na * Na2 + lambda_Cl * Cl2 + lambda_OH * OH2)
b3 = kappa3 - (lambda_H * H3 + lambda_Na * Na3 + lambda_Cl * Cl3 + lambda_OH * OH3)

#   (A * [_H3A, _H2A, _A]^T = b)
A = np.array([
    [C0,       0.5*C0,   0],   #   1: (_H3A + 0.5_H2A) * C0
    [0,        0.5*C0,   0],   #   2: 0.5_H2A * C0
    [0,          0,    0.5*C0]  #   3: 0.5_A * C0
])

#  
solution = np.linalg.solve(A, [b1, b2, b3])

# 
lambda_H3A, lambda_H2A, lambda_A = solution
print(f"(HA) = {lambda_H3A:.2f} /")
print(f"(HA)  = {lambda_H2A:.2f} /")
print(f"(A)    = {lambda_A:.2f} /")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# 
k1 = 2 * 10**(-2)
k2 = 9.12 * 10**(-7)
k3 = 8.128 * 10**(-10)

# ,  
def equation(y, x):
    term1 = x / (70 + x)
    term2 = (10**(-14) / y) - y
    numerator = 2 * k1 * k2 * k3 + k1 * k2 * y - y**3
    denominator = k1 * k2 * k3 + k1 * k2 * y + k1 * y**2 + y**3
    term3 = 0.01 * (numerator / denominator)
    return term1 - (term2 + term3)

#  x
x_values = np.linspace(-3, 4, 500)

#  y   x
y_values = []
for x in x_values:
    #    y
    y_initial_guess = 1e-12
    #   
    y = fsolve(equation, y_initial_guess, args=(x,))[0]
    y_values.append(y)

#  -lg(y)
neg_log_y_values = [-np.log10(y) for y in y_values]

plt.figure(figsize=(10, 6))
plt.plot(x_values, neg_log_y_values, label=" ")
plt.axhline(y=1.7, color='r', linestyle='--', label="pKa1")  #   y = 0.5
plt.axhline(y=6.04, color='r', linestyle='--', label="pKa2")  #   y = 0.5
plt.axhline(y=9.09, color='r', linestyle='--', label="pKa3")  #   y = -0.5

plt.axhline(y=4.12, color='g', linestyle='--', label=". 1")  #   y = 0.5
plt.axhline(y=7.58, color='g', linestyle='--', label="..2")  #   y = 0.5
plt.axhline(y=10.52, color='g', linestyle='--', label="..3")  #   y = -0.5


plt.title("  ")
plt.xlabel("V, ")
plt.ylabel("pH")
plt.grid(True)
plt.legend()
plt.show()
```
