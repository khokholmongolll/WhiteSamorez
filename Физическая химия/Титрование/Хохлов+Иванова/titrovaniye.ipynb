```python
import numpy as np
import matplotlib.pyplot as plt

# Константы
V_histidine = 70  # мл, объем гистидина
C_histidine = 0.01  # М, концентрация гистидина
C_NaOH = 1  # М, концентрация NaOH

# pKa значения для гистидина
pKa1 = 1.8  # карбоксильная группа
pKa2 = 6.0  # имидазольная группа
pKa3 = 9.2  # аминогруппа

# Количество вещества гистидина
n_histidine = V_histidine * C_histidine / 1000  # моль

# Объемы NaOH для титрования
V_NaOH = np.linspace(0, 10, 500)  # мл, от 0 до 10 мл с шагом

# Функция для расчета pH
def calculate_pH(V_NaOH):
    n_NaOH = V_NaOH * C_NaOH / 1000  # моль NaOH
    if n_NaOH < n_histidine:
        # Первая ступень титрования (карбоксильная группа)
        ratio = n_NaOH / n_histidine
        pH = pKa1 + np.log10(ratio / (1 - ratio))
    elif n_NaOH < 2 * n_histidine:
        # Вторая ступень титрования (имидазольная группа)
        ratio = (n_NaOH - n_histidine) / n_histidine
        pH = pKa2 + np.log10(ratio / (1 - ratio))
    else:
        # Третья ступень титрования (аминогруппа)
        ratio = (n_NaOH - 2 * n_histidine) / n_histidine
        pH = pKa3 + np.log10(ratio / (1 - ratio))
    return pH

# Рассчитаем pH для каждого объема NaOH
pH_values = [calculate_pH(V) for V in V_NaOH]

# Построим график
plt.figure(figsize=(10, 6))
plt.plot(V_NaOH, pH_values, label="Кривая титрования гистидина NaOH")
plt.title("Теоретическая кривая титрования гистидина NaOH")
plt.xlabel("Объем NaOH, мл")
plt.ylabel("pH")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Константы диссоциации
K1 = 10**(-1.8)   # ≈ 10^(-1.8)
K2 = 10**(-6.0)    # ≈ 10^(-6.0)
K3 = 10**(-9.2)    # ≈ 10^(-9.2)

C0 = 0.01          # М
V0 = 70            # мл

# Диапазон объемов NaOH от -3 до 4 мл
V_NaOH = np.linspace(-3, 4, 500)

# Функция для расчета [H+]
def calculate_H(V_NaOH):
    # Обработка отрицательных объемов (n_OH не может быть < 0)
    n_OH = max(V_NaOH * 1, 0)  # ммоль (1M NaOH)
    V_total = V0 + max(V_NaOH, 0)  # Объем не может быть меньше V0

    def equation(H):
        H = H[0]
        denominator = K1*K2*K3 + K1*K2*H + K1*H**2 + H**3
        n_H3A = (C0 * H**3 / denominator) * V_total
        n_H2A = (C0 * K1 * H**2 / denominator) * V_total
        n_HA = (C0 * K1*K2 * H / denominator) * V_total
        n_A = (C0 * K1*K2*K3 / denominator) * V_total
        return (n_H3A*2 + n_H2A*1 + n_HA*0 - n_OH) - (1e-3 * V_total * H)

    H_initial = 1e-7 if V_NaOH >= 0 else 1e-1  # Начальное приближение для H+
    H_solution = fsolve(equation, H_initial)
    return H_solution[0]

# Рассчитаем [H+] для каждого объема NaOH
H_values = np.array([calculate_H(V) for V in V_NaOH])

# Концентрации форм гистидина
denominator = K1*K2*K3 + K1*K2*H_values + K1*H_values**2 + H_values**3
A = (C0 * K1*K2*K3) / denominator
HA = (C0 * K1*K2 * H_values) / denominator
H2A = (C0 * K1 * H_values**2) / denominator
H3A = (C0 * H_values**3) / denominator

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(V_NaOH, H3A, label="[H₃A²⁺]", color="darkblue")
plt.plot(V_NaOH, H2A, label="[H₂A⁺]", color="blue")
plt.plot(V_NaOH, HA, label="[HA]", color="lightblue")
plt.plot(V_NaOH, A, label="[A⁻]", color="cyan")

plt.title("Зависимость концентрации форм гистидина от объема NaOH")
plt.xlabel("Объем NaOH, мл")
plt.ylabel("Концентрация, М")
plt.legend()
plt.grid(True)
plt.xlim(-3, 4)
plt.ylim(0, C0)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import math

x=np.array([ 2.05, 2.25, 2.45, 2.65, 2.95, 3.25, 3.85])
y=np.array([ 3360, 3930, 4470, 5050, 5880, 6720, 8320])

def least_squares(x, y):
    n = len(x)
    sumx = sum(x)/n
    sumy = sum(y)/n
    sumxy = sum(x*y)/n
    sumx_sq = sum(x**2)/n
    sumy_sq = sum(y**2)/n
    
    a = (sumxy - sumx * sumy) / (sumx_sq - sumx**2)
    b = sumy - a * sumx
    
    variance_a = (1/n**(1/2)) * (((sumy_sq - sumy**2)/(sumx_sq - sumx**2)) - b**2)**(1/2)
    variance_b = variance_a * (sumx_sq - sumx**2)**(1/2)
    
    return a, b, variance_a, variance_b

a, b, variance_a, variance_b = least_squares(x, y)
plt.scatter(x, y, color='blue')
plt.plot(x, a*x + b, color='red')
plt.xlabel('V, мл')
plt.ylabel('kappa, мСм/см')
plt.grid(True)
plt.title('' )
plt.show()

print(a,b, variance_a)
```

```python
import numpy as np

# Известные константы (в См·см²/моль)
lambda_Na = 50.1
lambda_Cl = 76.3
lambda_H = 349.65
lambda_OH = 147.16  # Найдено ранее

# Экспериментальные данные (в См/см)
kappa1 = 10900e-6  # 0.0109 См/см
kappa2 = 1145e-6   # 0.001145 См/см
kappa3 = 1607e-6   # 0.001607 См/см

# Концентрации
C0 = 0.01  # М (исходная концентрация гистидина)
V0 = 70     # мл (объем раствора)
C1 = 1.0    # М (концентрация титранта)

# Объемы титранта для каждой точки
V1 = -1.05  # мл (добавление HCl)
V2 = 0.38   # мл (добавление NaOH)
V3 = 1.06   # мл (добавление NaOH)

# Расчет [H+], [Cl-], [Na+], [OH-] для каждой точки
def calculate_concentrations(Vt):
    if Vt < 0:
        # Добавление HCl: [Cl-] = |Vt|*C1 / V0, [Na+] = 0
        Cl = abs(Vt) * C1 / V0
        Na = 0.0
        OH = 0.0  # В кислой среде [OH-] пренебрежимо мала
    else:
        # Добавление NaOH: [Na+] = Vt*C1 / V0, [Cl-] = 0
        Na = Vt * C1 / V0
        Cl = 0.0
        OH = Na  # В щелочной среде [OH-] ≈ [Na+]
    return Na, Cl, OH

# Для точки 1 (V1 = -1.05 мл, pH1 = 1.72)
Na1, Cl1, OH1 = calculate_concentrations(V1)
H1 = 10**(-1.72)  # [H+] = 10^(-pH)

# Для точки 2 (V2 = 0.38 мл, pH2 = 6.05)
Na2, Cl2, OH2 = calculate_concentrations(V2)
H2 = 10**(-6.05)

# Для точки 3 (V3 = 1.06 мл, pH3 = 9.07)
Na3, Cl3, OH3 = calculate_concentrations(V3)
H3 = 10**(-9.07)

# Вектор свободных членов (b = κ - известные слагаемые)
b1 = kappa1 - (lambda_H * H1 + lambda_Cl * Cl1)
b2 = kappa2 - (lambda_H * H2 + lambda_Na * Na2 + lambda_Cl * Cl2 + lambda_OH * OH2)
b3 = kappa3 - (lambda_H * H3 + lambda_Na * Na3 + lambda_Cl * Cl3 + lambda_OH * OH3)

# Матрица коэффициентов (A * [λ_H3A, λ_H2A, λ_A]^T = b)
A = np.array([
    [C0,       0.5*C0,   0],   # Уравнение для κ1: (λ_H3A + 0.5λ_H2A) * C0
    [0,        0.5*C0,   0],   # Уравнение для κ2: 0.5λ_H2A * C0
    [0,          0,    0.5*C0]  # Уравнение для κ3: 0.5λ_A * C0
])

# Решение системы
solution = np.linalg.solve(A, [b1, b2, b3])

# Результаты
lambda_H3A, lambda_H2A, lambda_A = solution
print(f"λ⁰(H₃A²⁺) = {lambda_H3A:.2f} См·см²/моль")
print(f"λ⁰(H₂A⁺)  = {lambda_H2A:.2f} См·см²/моль")
print(f"λ⁰(A⁻)    = {lambda_A:.2f} См·см²/моль")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Константы
k1 = 2 * 10**(-2)
k2 = 9.12 * 10**(-7)
k3 = 8.128 * 10**(-10)

# Функция, представляющая уравнение
def equation(y, x):
    term1 = x / (70 + x)
    term2 = (10**(-14) / y) - y
    numerator = 2 * k1 * k2 * k3 + k1 * k2 * y - y**3
    denominator = k1 * k2 * k3 + k1 * k2 * y + k1 * y**2 + y**3
    term3 = 0.01 * (numerator / denominator)
    return term1 - (term2 + term3)

# Диапазон x
x_values = np.linspace(-3, 4, 500)

# Найдем y для каждого x
y_values = []
for x in x_values:
    # Начальное предположение для y
    y_initial_guess = 1e-12
    # Решаем уравнение численно
    y = fsolve(equation, y_initial_guess, args=(x,))[0]
    y_values.append(y)

# Вычисляем -lg(y)
neg_log_y_values = [-np.log10(y) for y in y_values]

plt.figure(figsize=(10, 6))
plt.plot(x_values, neg_log_y_values, label="Теоретическая кривая")
plt.axhline(y=1.7, color='r', linestyle='--', label="pKa1")  # Линия на y = 0.5
plt.axhline(y=6.04, color='r', linestyle='--', label="pKa2")  # Линия на y = 0.5
plt.axhline(y=9.09, color='r', linestyle='--', label="pKa3")  # Линия на y = -0.5

plt.axhline(y=4.12, color='g', linestyle='--', label="т.э 1")  # Линия на y = 0.5
plt.axhline(y=7.58, color='g', linestyle='--', label="т.э.2")  # Линия на y = 0.5
plt.axhline(y=10.52, color='g', linestyle='--', label="т.э.3")  # Линия на y = -0.5


plt.title("Теоретическая кривая титрования")
plt.xlabel("V, мл")
plt.ylabel("pH")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Пример массивов X и Z
X = np.array([-2.5, -1.9, -1.3, -1.1, -0.9, -0.7, -0.5, -0.4, -0.3, -0.25, -0.2, -0.15, -0.12, -0.09, -0.06, -0.05, 0.0, 0.03, 0.05, 0.07, 0.09, 0.11, 0.13, 0.15, 0.17, 0.2, 0.22, 0.24, 0.26, 0.28, 0.3, 0.32, 0.35, 0.38, 0.41, 0.46, 0.51, 0.56, 0.61, 0.66, 0.71, 0.76, 0.79, 0.82, 0.85, 0.88, 0.91, 0.94, 0.97, 1.0, 1.05, 1.1, 1.15, 1.2, 1.25, 1.3, 1.35, 1.4, 1.45, 1.5, 1.55, 1.6, 1.65, 1.7, 1.75, 1.8, 1.85, 1.95, 2.05, 2.25, 2.45, 2.65, 2.95, 3.25, 3.85])
Z = np.array([1.62, 1.75, 1.92, 1.98, 2.091, 2.204, 2.356, 2.45, 2.58, 2.66, 2.76, 2.885, 2.992, 3.111, 3.283, 3.367, 4.11, 4.68, 4.93, 5.094, 5.178, 5.274, 5.356, 5.433, 5.506, 5.595, 5.65, 5.705, 5.76, 5.8, 5.85, 5.892, 5.963, 6.03, 6.1, 6.209, 6.272, 6.41, 6.55, 6.736, 6.965, 7.35, 7.67, 7.98, 8.204, 8.35, 8.49, 8.58, 8.7, 8.765, 8.915, 9.0, 9.12, 9.243, 9.36, 9.472, 9.57, 9.72, 9.88, 10.01, 10.223, 10.47, 10.75, 10.956, 11.095, 11.215, 11.29, 11.4, 11.517, 11.65, 11.731, 11.8, 11.8, 11.93, 12.015])

# Вычисляем разности между соседними элементами
delta_X = X[:-1] - X[1:]  # X_{n-1} - X_n
delta_Z = Z[:-1] - Z[1:]  # Z_{n-1} - Z_n

# Вычисляем значение (X_{n-1} - X_n) / (Z_{n-1} - Z_n)
ratio = delta_Z / delta_X

# Ищем локальные минимумы
# Используем -ratio, так как find_peaks ищет максимумы
minima_indices, _ = find_peaks(-ratio, prominence=0.1)  # prominence задает минимальную значимость минимума

# Соответствующие значения Z и X для локальных минимумов
Z_minima = Z[minima_indices + 1]  # +1, так как ratio на один элемент короче
X_minima = X[minima_indices + 1]

# Построим график
plt.figure(figsize=(10, 6))
plt.scatter(Z[1:], ratio, color='blue', label="Экспериментальная зависимость")
plt.plot(Z[1:], ratio, color='blue')

# Отмечаем локальные минимумы на графике
plt.scatter(Z_minima, ratio[minima_indices], color='red', label="Локальные минимумы")

# Подписываем точки минимумов
for i, (x, z) in enumerate(zip(X_minima, Z_minima)):
    plt.text(z, ratio[minima_indices[i]], f"({x:.2f}, {z:.2f})", fontsize=9, ha='right')

plt.title("График разностных производных с локальными минимумами")
plt.xlabel("pH")
plt.ylabel("dpH/dV")
plt.grid(True)
plt.legend()
plt.show()

# Выводим локальные минимумы
print("Локальные минимумы:")
for x, z in zip(X_minima, Z_minima):
    print(f"X = {x:.2f}, Z = {z:.2f}")
```
